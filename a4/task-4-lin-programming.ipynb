```python
import numpy as np
import scipy.optimize as opt
```

### No Standard Form


```python
x = np.ones(30)
costs_1 =  [0.5, 0.25, 0.25, 0.25, 1.0, 1.0, 0.5, 0.5, 1.0, 0.5, 1.5, 2.5, 1.0, 2.5, 3.5]
costs_2 = [0.75, 1.0, 0.75, 0.5, 0.5, 0.25, 0.25, 0.25, 2.0, 1.25, 1.0, 4.0, 2.5, 3.0, 2.0]

c = [0.5, 0.25, 0.25, 0.25, 1.0, 1.0, 0.5, 0.5, 1.0, 0.5, 1.5, 2.5, 1.0, 2.5, 3.5,
     0.75, 1.0, 0.75, 0.5, 0.5, 0.25, 0.25, 0.25, 2.0, 1.25, 1.0, 4.0, 2.5, 3.0, 2.0,]
```

```python
all_tasks_constraint = [np.zeros(30), np.zeros(30)]
b_eq = [0 ,0]
for i in range(15):
    arr = np.zeros(30)
    arr[i] = 1
    arr[i + 15] = 1
    all_tasks_constraint.append(arr)
    b_eq.append(1)

all_tasks_constraint = np.array(all_tasks_constraint)

print(all_tasks_constraint.shape)

```

```python
inequality_constraint = [np.append(costs_1, np.zeros(15)), np.append(np.zeros(15), costs_2)]
b_ineq = [9, 6]
for i in range(15):
    arr = np.zeros(30)
    inequality_constraint.append(arr)
    b_ineq.append(0)

    
inequality_constraint = np.array(inequality_constraint)
print(inequality_constraint.shape)
```

```python
A_eq = all_tasks_constraint
A_ineq = inequality_constraint
bounds = [(0, 1) for _ in range(30)]
print(A_eq)
print("==================")
print(A_ineq)
print("==================")
print(b_eq)
print("==================")
print(b_ineq)
print("==================")
```

```python
sol = opt.linprog(c, A_ub=A_ineq, b_ub=b_ineq, A_eq=A_eq, b_eq=b_eq, x0=x,integrality=1, bounds=bounds)['x']
print(sol[:15], np.sum(sol[:15]@costs_1))
print(sol[15:], np.sum(sol[15:]@costs_2))
print(np.sum(sol@c))
```

### Standard Form


```python
costs_1_s =  [0.5, 0.25, 0.25, 0.25, 1.0, 1.0, 0.5, 0.5, 1.0, 0.5, 1.5, 2.5, 1.0, 2.5, 3.5, 9.0]
costs_2_s = [0.75, 1.0, 0.75, 0.5, 0.5, 0.25, 0.25, 0.25, 2.0, 1.25, 1.0, 4.0, 2.5, 3.0, 2.0, 6.0]

c_s = [0.5, 0.25, 0.25, 0.25, 1.0, 1.0, 0.5, 0.5, 1.0, 0.5, 1.5, 2.5, 1.0, 2.5, 3.5, 0.0,
     0.75, 1.0, 0.75, 0.5, 0.5, 0.25, 0.25, 0.25, 2.0, 1.25, 1.0, 4.0, 2.5, 3.0, 2.0, 0.0]
```

```python
A_eq_s = [np.append(costs_1_s, np.zeros(16)), np.append(np.zeros(16), costs_2_s)]
b_eq_s = [9 ,6]
for i in range(15):
    arr = np.zeros(32)
    arr[i] = 1
    arr[i + 16] = 1
    A_eq_s.append(arr)
    b_eq_s.append(1)

A_eq_s = np.array(A_eq_s)
print(b_eq_s)

print(A_eq_s.shape)
```

```python
import sys

bounds_s = [(0, 1) for _ in range(32)]
bounds_s[15]=(0, sys.maxsize)
bounds_s[31]=(0, sys.maxsize)
```

```python
sol_s = opt.linprog(c_s, A_eq=A_eq_s, b_eq=b_eq_s, bounds=bounds_s)['x']
s_1 = np.sum(sol_s[:15]@costs_1_s[:15])
print(s_1)
s_2 = np.sum(sol_s[16:31]@costs_2_s[:15])
print(s_2)
print(s_1 + s_2)
```
